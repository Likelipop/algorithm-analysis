## Bài 1

Một mảng $A$ có $n$ phần tử. Ta gọi $x \in A$ là phần tử **"lấn chiếm cấp độ 10"** nếu như $x$ xuất hiện ít nhất $\left\lceil \frac{n}{10} \right\rceil$ lần trong $A$.


### 1. Xây dựng một thuật toán tìm phần tử lấn chiếm cấp 10 trong một mảng.

---


```python
import random
import math
import time

# reusable snippet for generating random arrays
def generate_random_array(size, min_val=1, max_val=20):
    return [random.randint(min_val, max_val) for _ in range(size)]

# ==========================================
# PROBLEM 1: ALGORITHM EXECUTION
# ==========================================
print("--- PROBLEM 1: ALGORITHM EXECUTION ---")

test_cases = [
    generate_random_array(20, 1, 5),
    generate_random_array(50, 1, 10),
    generate_random_array(100, 1, 15)
]

for tc_idx, original_arr in enumerate(test_cases):
    # SET UP TEST CASE ĐỂ CHẠY
    arr = original_arr.copy()
    n_original = len(arr)
    threshold = math.ceil(n_original / 10)

    result_elements = []
    # track the effective number of elements as 'None' values are pushed
    current_n = n_original

    print(f"test_case_{tc_idx + 1} (size {n_original}):")
    print(f"original_array: {original_arr}")
    print(f"original_n: {n_original}, threshold: {threshold}")
    if n_original == 20:
        print("-" * 50)


    # THUẬT TOÁN BẮT ĐẦU TỪ ĐÂY
    i = 0
    while i < n_original:
        if arr[i] is None:
            i += 1
            continue

        j = i + 1
        d = 0

        while j < n_original:
            if arr[j] is not None and arr[j] == arr[i]:
                d += 1
                if j + d < n_original:
                    arr[j] = arr[j + d]
                    arr[j + d] = None
                else:
                    arr[j] = None
                current_n -= 1
            else:
                j += 1

        if (d + 1) >= threshold:
            result_elements.append(arr[i])

        # MINH HỌA TỪNG BƯỚC VỚI I
        if n_original == 20:
            shrunk_arr = [val for val in arr if val is not None]
            print(f"step i={i}, checking element '{arr[i]}':")
            print(f"  raw_array    : {arr}")
            print(f"  shrunk_array : {shrunk_arr}")
            print(f"  current_n    : {current_n}")
            print(f"  count        : {d + 1}")
            print("-" * 30)

        i += 1

    print(f"results: {result_elements}\n")
    if n_original == 20:
        print("=" * 50 + "\n")

```

Bên cạnh thuật toán N^2 thường thấy, em tạo ra thuật toán trên (mặc dù worse case nó vẫn là N^2 nhưng nó có thể linh hoạt theo tùy các trường hợp khác nhau về giá trị, thuật toán có thể kết thúc sớm hơn nhờ vào việc nó chỉ duyệt các trường hợp khác nhau duy nhất trên mỗi lần duyệt,...).

Đầu tiên là độ phức tạp, xét trường hợp tệ nhất, khi không có phần tử trùng, thời gian là $O(N^2)$. Ngược lại, với trường hợp ok nhất, việc thu hẹp mảng sẽ thực hiện ngay phần đầu tiên, dẫn tới thới gian là $O(1)$

### 2. Liệu có tồn tại thuật toán có độ phức tạp \( O(n) \) để tìm ra phần tử kiểu này không?


Để đạt thời gian $O(N)$, hướng tiếp cận đầu tiên là dùng Hash Map đếm tần suất. Tuy nhiên, do đề bài không quy định rõ kiểu dữ liệu của phần tử, việc thiết kế hàm băm có thể gặp rủi ro, khiến cấu trúc dữ liệu này trở nên khá mơ hồ và khó định hình toàn diện.

Cách thứ hai giúp tối ưu không gian lưu trữ xuống mức $O(1)$ là thuật toán Misra-Gries. Dù xét trong trường hợp tệ nhất, độ phức tạp thời gian luôn đảm bảo tuyến tính $O(N)$ và độ phức tạp bộ nhớ luôn được giới hạn ở hằng số cực kỳ tiết kiệm, không phụ thuộc vào kích thước mảng.

Ý tưởng cốt lõi của thuật toán gồm 2 bước:

1. Tìm ứng viên: Duyệt mảng, duy trì tối đa 9 ứng cử viên. Nếu gặp phần tử mới, trừ đi 1 lượt đếm của toàn bộ ứng viên. Việc trừ ứng viên này dựa trên một kết quả :": chọn ra một nhóm gồm đúng $k$ phần tử có giá trị đôi một khác nhau và loại bỏ nhóm đó khỏi mảng, thì thao tác này có thể thực hiện tối đa (N/k) lần",
vì vậy kết quả của bài toán sẽ là những phần tử còn ở mảng tạm (sau khi duyệt xong)
2. Xác thực: Duyệt mảng lần nữa để đếm lại tần suất thực tế của các ứng cử viên và chốt kết quả.

Điểm tương đồng là cả hai đều đếm và lưu trữ các cặp "phần tử - tần suất". Sự khác biệt nằm ở quản lý bộ nhớ: Hash Map lưu mọi giá trị nên tốn không gian $O(N)$. Ngược lại, Misra-Gries dùng cơ chế loại trừ  để giới hạn đúng 9 ứng viên, giúp bộ nhớ luôn cố định $O(1)$.



```python
import random
import math
import time

def generate_random_array(size, min_val=1, max_val=20):
    return [random.randint(min_val, max_val) for _ in range(size)]

# ==========================================
# GIẢI THUẬT
# ==========================================
print("--- PROBLEM 1: MISRA-GRIES O(N) EXECUTION ---")

test_cases = [
    generate_random_array(20, 1, 5),
    generate_random_array(50, 1, 10),
    generate_random_array(100, 1, 15)
]

for tc_idx, original_arr in enumerate(test_cases):
    arr = original_arr.copy()

    n = len(arr)
    threshold = math.ceil(n / 10)
    max_candidates = 10

    # candidates[i] represents the value of the i-th candidate.
    candidates = []

    # counts[i] represents the frequency vote of candidates[i].
    counts = []

    for num in arr:
        if num in candidates:
            idx = candidates.index(num)
            counts[idx] += 1
        elif len(candidates) < max_candidates:
            candidates.append(num)
            counts.append(1)
        else:
            idx = 0
            while idx < len(counts):
                counts[idx] -= 1
                if counts[idx] == 0:
                    candidates.pop(idx)
                    counts.pop(idx)
                else:
                    idx += 1

    result_elements = []

    for candidate in candidates:
        actual_count = 0
        for num in arr:
            if num == candidate:
                actual_count += 1

        if actual_count >= threshold:
            result_elements.append(candidate)

    print(f"test_case_{tc_idx + 1}:")
    print(f"original_array: {original_arr}")
    print(f"n_elements: {n}, threshold: {threshold}")
    print(f"results: {result_elements}\n")


```

## Bài 2

Cho mảng \( A \) và \( B \) đã được sắp xếp theo thứ tự tăng dần và có \( n \) phần tử.

Hãy xây dựng một thuật toán khả thi để **merge** hai mảng \( A \) và \( B \) thành mảng có \( 2n \) phần tử đã được sắp xếp theo thứ tự tăng dần.

Bài này, do hai mảng sắp xếp rồi, ta sẽ lợi dụng tính chất này và dùng hai con trỏ để duyệt với độ phức tạp O(N).

```python
n_size = 10
array_a = sorted(generate_random_array(n_size, 1, 100))
array_b = sorted(generate_random_array(n_size, 1, 100))

merged_array = []
ptr_a = 0
ptr_b = 0
len_a = len(array_a)
len_b = len(array_b)

while ptr_a < len_a and ptr_b < len_b:
    if array_a[ptr_a] <= array_b[ptr_b]:
        merged_array.append(array_a[ptr_a])
        ptr_a += 1
    else:
        merged_array.append(array_b[ptr_b])
        ptr_b += 1

while ptr_a < len_a:
    merged_array.append(array_a[ptr_a])
    ptr_a += 1

while ptr_b < len_b:
    merged_array.append(array_b[ptr_b])
    ptr_b += 1

print(f"array_a: {array_a}")
print(f"array_b: {array_b}")
print(f"merged_array: {merged_array}")
```
